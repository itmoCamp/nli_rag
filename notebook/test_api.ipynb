```python
import requests
def test_nda(text: str, task: str = "nda"):
    """
    Отправляет POST-запрос к локальному сервису на таск 'nda' и выводит результат.
    """
    url = "http://localhost:8001/answer"
    payload = {
        "text": text,
        "task": task
    }
    headers = {
        "Content-Type": "application/json"
    }
    response = requests.post(url, json=payload, headers=headers)
    try:
        response.raise_for_status()
    except requests.HTTPError:
        print(f"Ошибка {response.status_code}: {response.text}")
        return

    data = response.json()
    print("Ответ сервиса:")
    for item in data:
        print(f"- Гипотеза: {item['hypothesis']}")
        print(f"  Метка: {item['label']} (код: {item['label_num']})")
        print("  Reasoning:", item.get('reasoning', '—'))
        print("  Quote:", item.get('quote', '—'))
        print("  Фрагменты:")
        for chunk in item.get('chunks', []):
            print(f"    - {chunk}")
        print(f"  Span: {item['start']}–{item['end']}")
        print()
```
